```python
# Recursive functions v iterated functions run in different directions. 
# Iterated Functions: Outside --> In
# Recursive Functions: Inside --> Out
```

```python
nested_list = [[1, 2, 3], ["a", "b", "c"], [True, False, None]]

for sublist in nested_list:
    for element in sublist:
        print(element)
```

```python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    else:
        return n * factorial(n - 1)

result = factorial(5)
print(result)  
```
